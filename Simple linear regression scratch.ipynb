```python
"""
Linear regression is technique where we model relationship between dependent and independent variable.
here,
equation of line :
Y = M * X + C
X = independent variable
Y = dependent variable
M = slope of line
C = starting point of regression line or y intercept
"""

indep = [1,2,3,4,5,6,7,8,9]
dep = [21,34,67,68,64,77,77,89,112]
import matplotlib.pyplot as plt
plt.scatter(indep,dep, marker="o")
plt.show()
```

```python
import numpy as np
import statistics as st

dep_mean = st.mean(dep)
indep_mean = st.mean(indep)
plt.plot(indep,dep)

plt.axvline(x = indep_mean, color='r',linestyle = '--')
plt.axhline(y = dep_mean, color = 'r', linestyle = '--')
plt.title("Mean and current distribution",fontdict = font1)   
plt.xlabel("Independant Variable")
plt.ylabel("Dependant Variable")
```

```python
# Finding errors and best fit line 

diff_x = []
position = 0
while position < len(indep):
    diff_x.append((indep[position] - indep_mean))
    position +=1
    
diff_x_sq = []
position = 0
while position < len(indep):
    diff_x_sq.append((indep[position] - indep_mean)**2)
    position +=1
    
diff_x_sq = np.sum(diff_x_sq)

diff_y = []
position = 0
while position < len(dep):
    diff_y.append((dep[position] - dep_mean))
    position +=1  
    
position  = 0
sum_error = []
while position < len(diff_x):
    sum_error.append(diff_x[position] * diff_y[position])
    position +=1

sum_error_total = np.sum(sum_error)
M = sum_error_total/diff_x_sq
C = dep_mean - (M * indep_mean)
new_y = []
position = 0
while position < len(indep):
    new_y.append(M * indep[position] + C)
    position += 1
plt.scatter(indep,dep, marker="o")
plt.plot(indep,new_y,label="line1")
font1 = {'family':'serif','color':'Red','size':20}

plt.title("New Regression line ",fontdict = font1)   
plt.xlabel("Independant Variable")
plt.ylabel("Dependant Variable")
```

```python
########################R-SQUARE VALUE#################################
diff_y_square = []
position = 0
while position < len(dep):
    diff_y_square.append((dep[position] - dep_mean)**2)
    position +=1 
diff_y_square = np.sum(diff_y_square)


diff_y_square_predicted = []
position = 0
while position < len(new_y):
    diff_y_square_predicted.append((new_y[position] - dep_mean)**2)
    position +=1 
diff_y_square_predicted = np.sum(diff_y_square_predicted)


print("Original variance:",diff_y_square)
print("Variance explained by model:",diff_y_square_predicted)

R_square = diff_y_square_predicted/diff_y_square
print("R-square = Predicted Variance/ Original variance")
print("R_Square:",R_square)
```
